## **2.3.1 Joint Probability Distributions**

**Joint Probability Distribution:** the probability of two or more events happening together. A joint probability distribution shows a probability distribution for two (or more) random variables

**Two Discrete Random Variables:** For discrete random variables ùëã and ùëå, a joint probability distribution assigns probabilities to each pair of outcomes.

```python
# An example of a probability table for joint distribution of two discrete random variables
joint_probability = pd.DataFrame({
    'X': ['A', 'B', 'A', 'B'],
    'Y': ['C', 'C', 'D', 'D'],
    'P(X,Y)': [0.1, 0.3, 0.4, 0.2]
})

joint_probability
```

**Two Continuous Random Variables:**

```python
# Create two continuous random variables X and Y
X = np.random.normal(0, 1, 1000)
Y = np.random.normal(0, 1, 1000)

# Use a scatter plot to demonstrate the joint distribution
plt.scatter(X, Y, alpha=0.5, edgecolor='none')
plt.title('Two Continuous Random Variables')
plt.xlabel('X Norm')
plt.ylabel('Y Norm')
plt.grid(True)
plt.show()
```

**Independent Random Variables:** Two random variables ùëã and ùëå are independent if ùëÉ(ùëã,ùëå) = ùëÉ(ùëã)ùëÉ(ùëå) . This shows that the occurrence of one event doesn't affect the other.

## **2.3.2. Correlation and Dependence**

**Correlation and Dependence:** Correlation measures the strength of a linear relationship between two random variables. Dependence refers to any kind of relationship between them - not needing to be linear.

**Correlation for Random Variables:** Correlation is usually quantified using Pearson's correlation coefficient: image.png

```python
# Create two random variables X and Y
X = np.random.normal(0, 1, 1000)
Y = 0.5 * X + np.random.normal(0, 1, 1000)

# Calculate the Pearson correlation for (X,Y)
correlation = np.corrcoef(X, Y)[0, 1]
print(f"Pearson Correlation Coefficient for (X,Y): {correlation:.3f}")
```

## **2.3.3. Random Samples**

**Random Samples:** A random sample is a subset of individuals chosen from a larger set (population) where each individual is chosen randomly and purely by chance.

```python
# Sample set
sample_set = np.arange(1, 101)

# Take 5 random samples from our set
random_samples = np.random.choice(sample_set, size=5, replace=False)
print("The 5 random samples from the original set from 1-100:", random_samples)
```

**The Central Limit Theorem (CLT):** the distribution of sample means approximates a normal distribution as the sample size gets larger, regardless of the population's distribution.

```python
# Function that takes the number of samples and the number of experiments to
# perform and generates multiple sample means by drawing random samples from a
# specified distribution and return the list of those means
def simulate_sample_means(n, num_samples):
    sample_means = []
    for _ in range(num_samples):
        sample = np.random.exponential(scale=2, size=n)
        sample_means.append(np.mean(sample))
    return sample_means

# Using 3 different sizes to show the similarity
sample_means_10 = simulate_sample_means(10, 1000)
sample_means_50 = simulate_sample_means(50, 1000)
sample_means_100 = simulate_sample_means(100, 1000)

# Create a historgram to show no matter the size the CLT will approximate a normal distribution
plt.hist(sample_means_10, bins=30, alpha=0.5, label='n=10')
plt.hist(sample_means_50, bins=30, alpha=0.5, label='n=50')
plt.hist(sample_means_100, bins=30, alpha=0.5, label='n=100')
plt.legend(loc='upper right')
plt.title("Central Limit Theorem Demonstration")
plt.show()
```
